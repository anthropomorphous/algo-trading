```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Tesla_Stock.csv', index_col = 'Date', parse_dates=True)
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df.rolling(window=7).mean().head(10)
```

```python
# each value now represents its last 7 days average

df['Open'].plot()
df.rolling(window=7).mean()['Close'].plot(figsize=(16,6))
```

```python
# MA - Moving Average

df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA', 'Close']].plot(figsize=(16,6))
```

```python
df['Close'].expanding().mean().plot(figsize=(16,6))
```

# Bollinger Bands

* Close 20 MA
* Upper = 20 MA + 2*std(20)
* Lower = 20 MA - 2*std(20)
* Close

```python
# CLose 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

# Upper 
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())

# Lower
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())

# Close
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].tail(365).plot(figsize=(16,6))
```
