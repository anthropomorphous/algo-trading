# Daily Percentage Change

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pandas_datareader
import datetime
import pandas_datareader.data as web

%matplotlib inline
```

```python
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2020,1,1)
```

```python
tesla = web.DataReader('TSLA', 'yahoo', start,end)
ford = web.DataReader('F', 'yahoo', start, end)
genmot = web.DataReader('GM', 'yahoo', start, end)
```

```python
tesla['returns'] = (tesla['Close'] / tesla['Close'].shift(1)) - 1
```

```python
# or
tesla['returns'] = tesla['Close'].pct_change(1)

ford['returns'] = ford['Close'].pct_change(1)
genmot['returns'] = genmot['Close'].pct_change(1)
```

```python
tesla['returns']
```

```python
# What stock is more volatile?
```

```python
ford['returns'].hist(bins=100)
```

```python
genmot['returns'].hist(bins=100)
```

```python
tesla['returns'].hist(bins=100)
```

```python
tesla['returns'].hist(bins=100, label='Tesla', figsize=(10,8), alpha=0.5)
genmot['returns'].hist(bins=100, label='General Motors', figsize=(10,8), alpha=0.5)
ford['returns'].hist(bins=100, label='Ford', figsize=(10,8), alpha=0.5)
plt.legend();
```

```python
tesla['returns'].plot(kind='kde', label='Tesla', figsize=(10,8))
genmot['returns'].plot(kind='kde', label='General Motors', figsize=(10,8))
ford['returns'].plot(kind='kde', label='Ford', figsize=(10,8))

plt.legend();
```

```python
# Here we can see (somehow) that Ford is the most stable and Tesla is the most volatile
```

```python
box_df = pd.concat([tesla['returns'], ford['returns'], genmot['returns']],axis=1)
box_df.columns = ['Tesla Ret', 'Ford Ret', 'GM Ret']
box_df.plot(kind='box', figsize=(8,11))
```

```python
# I.e. Tesla returns have much wider distribution
```

# Comparing Daily Returns between Stocks

```python
from pandas.plotting import scatter_matrix

scatter_matrix(box_df, figsize=(8,8), alpha=0.2, hist_kwds={'bins':100})
```

```python
box_df.plot(kind='scatter',x='Ford Ret', y='GM Ret', alpha=0.5, figsize=(8,8))
```

```python
# Here we can see that this two stocks are correlating with each other,
# as their relationship we can define as pretty much as linear model
```

# Cumulative Daily Returns

```python
# is an aggregate amount an investment has gained or lost over time,
# independent of the period of time involved
```

```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
```

```python
tesla
```

```python
# i.e. if we have invested in the first day (2012-01-03) our returns on the
# second and next days will be in minus (a loss)
# but when Cumulative Returns became higher that 1, we start to have ROI
```

```python
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
genmot['Cumulative Return'] = (1 + genmot['returns']).cumprod()
```

```python
tesla['Cumulative Return'].plot(label='Tesla', figsize=(16,8))
ford['Cumulative Return'].plot(label='Ford')
genmot['Cumulative Return'].plot(label='General Motors')
plt.legend();
```

```python
# or only comparing Ford and General Motors

ford['Cumulative Return'].plot(label='Ford', figsize=(16, 8))
genmot['Cumulative Return'].plot(label='General Motors')
plt.legend();
```
